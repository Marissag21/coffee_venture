```python
#import dependencies
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
#read the CSV == source - https://www.kaggle.com/datasets/patkle/coffeereviewcom-over-7000-ratings-and-reviews
csv_data = "reviews_feb_2023.csv"
coffee_df = pd.read_csv(csv_data)
coffee_df.head()
```

```python
#check for null values
coffee_df.isnull().sum()
```

```python
#drop the unneccessary columns 'with_milk','agtron','roaster','url' and high null value cols which are 'acidity_structure', 'bottom_line'
coffee1_df = coffee_df.drop(columns=['with_milk','agtron','roaster','url','acidity_structure', 'bottom_line'])
coffee1_df.head()
```

```python
#check if cols are dropped and detect which cols have null values
coffee1_df.isnull().sum()
```

```python
#removed columns and null values to show 4905 rows × 13 columns with no null values
coffee2_df = coffee1_df.dropna(subset=['aftertaste', 'aroma', 'body', 'flavor', 'blind_assessment',
                                       'coffee_origin', 'est_price', 'notes', 'roast_level', 'roaster_location'])
coffee2_df.head()
```

```python
#check if there are any null values
coffee2_df.isnull().sum()
```

```python
coffee3_df = coffee2_df['roaster_location'].str.split(',', n=1, expand=True)
coffee3_df.columns=['city',  'country']
coffee3_df.head()
```
