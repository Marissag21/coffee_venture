```python
#import dependencies
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
#read the CSV == source - https://www.kaggle.com/datasets/patkle/coffeereviewcom-over-7000-ratings-and-reviews
csv_data = "reviews_feb_2023.csv"
coffee_df = pd.read_csv(csv_data)
coffee_df.head()
```

```python
#check for null values
coffee_df.isnull().sum()
```

```python
coffee_df["acidity_structure"].value_counts()
```

```python
coffee_df["acidity_structure"].isnull().sum()
```

```python
#drop the unneccessary columns 'with_milk','agtron','roaster','url' and high null value cols which are 'acidity_structure', 'bottom_line'
coffee1_df = coffee_df.drop(columns=['with_milk','agtron','roaster','url', 'bottom_line', "review_date"])
coffee1_df.head()
coffee1_df.index      
```

```python
#check if cols are dropped and detect which cols have null values
coffee1_df.isnull().sum()
```

```python
#removed columns and null values to show 4905 rows × 13 columns with no null values
coffee2_df = coffee1_df.dropna(subset=['title', 'acidity_structure', 'aftertaste', 'aroma', 'body', 'flavor', 'blind_assessment',
                                       'coffee_origin', 'est_price', 'notes', 'roast_level', 'roaster_location'])
coffee2_df = coffee2_df.drop_duplicates()
coffee2_df.reset_index(drop=True)

```

```python
#check for datatypes for any null values and datatypes
display(coffee2_df.info())
```

```python

coffee3_df = coffee2_df.astype({"title":"string", "rating": float, "aroma": float, "body": float, "flavor": float, "blind_assessment": "string", "coffee_origin": "string", "est_price": "string", "notes": "string", "roast_level": "string", "roaster_location":"string"})
display(coffee3_df.info())
```

```python
coffee3_df
```

```python
coffee3_df["roaster_location"].value_counts()
```

```python
coffee3_df.isnull().sum()
```

```python
coffee3_df[["City", "State", "Country"]] = coffee3_df["roaster_location"].apply(lambda x: pd.Series(str(x).split(",")))
coffee3_df
#coffee3_df.reset_index(drop=True)
```

```python
coffee3_df.to_csv("coffee_venture1.csv")
```
